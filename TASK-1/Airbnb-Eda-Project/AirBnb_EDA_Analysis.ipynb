```python
new_df = new_df.drop(['id','last_review'],axis=1)
```

```python
new_df.shape
```

```python
new_df.head()
```

```python
new_df.fillna({'host_name':'absent','reviews_per_month':0},inplace = True)
```

```python
new_df.isnull().sum()
```

```python
new_df['neighbourhood'] = new_df['neighbourhood'].ffill()
new_df['neighbourhood'] = new_df['neighbourhood'].bfill()
```

```python
new_df.isnull().sum()
```

```python
new_df['price'] = new_df['price'].replace(r'[\$,]', '', regex=True).astype(float)
```

```python

new_df.describe()
```

```python
# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# 1. Distribution of Prices
plt.figure(figsize=(10, 6))
sns.histplot(new_df['price'], bins=50, kde=True)
plt.title('Distribution of Prices')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.show()

# 2. Neighbourhood Popularity
plt.figure(figsize=(12, 8))
sns.countplot(y='neighbourhood', data=new_df, order=new_df['neighbourhood'].value_counts().index)
plt.title('Number of Listings per Neighbourhood')
plt.xlabel('Number of Listings')
plt.ylabel('Neighbourhood')
plt.show()

# 3. Room Type Distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='room_type', data=new_df)
plt.title('Distribution of Room Types')
plt.xlabel('Room Type')
plt.ylabel('Number of Listings')
plt.show()

# 4. Price vs. Number of Reviews
plt.figure(figsize=(10, 6))
sns.scatterplot(x='number_of_reviews', y='price', data=new_df)
plt.title('Price vs. Number of Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Price ($)')
plt.show()

# 5. Availability of Listings
plt.figure(figsize=(10, 6))
sns.histplot(new_df['availability_365'], bins=50, kde=True)
plt.title('Availability of Listings (Days per Year)')
plt.xlabel('Availability (Days)')
plt.ylabel('Frequency')
plt.show()

# 6. Price by Room Type
plt.figure(figsize=(10, 6))
sns.boxplot(x='room_type', y='price', data=new_df)
plt.title('Price by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price ($)')
plt.show()
```

```python
# Identify non-numeric columns
non_numeric_columns = new_df.select_dtypes(exclude=[np.number]).columns

# Drop non-numeric columns from the DataFrame
numeric_df = new_df.drop(non_numeric_columns, axis=1)

# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()

```

```python
average_price = new_df['price'].mean()
print(f'Average price of listings: ${average_price:.2f}')
```

Analysis and Data Insights
1. Correlation Analysis
From the correlation matrix, we can observe the relationships between various features in the dataset. Here are some insights based on the heatmap:

 - Price Correlations: The price does not show strong correlation with most other features. The highest correlation (although still weak) is with latitude (-0.10), suggesting slight regional price variations.
 - Number of Reviews and Reviews per Month: The number_of_reviews and reviews_per_month show a strong positive correlation (0.60). This makes sense as more frequently reviewed listings will naturally accumulate more total reviews.
 - Host Listings Count: The calculated_host_listings_count shows weak correlations with other features, slightly positive with price (0.06) and slightly negative with availability_365 (-0.09).
 - Availability: The availability_365 shows a weak positive correlation with reviews_per_month (0.14), suggesting that listings with higher availability might receive more frequent reviews.
These insights suggest that most features have weak correlations with each other, indicating a diverse set of influences on listing characteristics and performance.
